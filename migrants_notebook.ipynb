    <p>Missing Migrants Project records since 2014 people who die in the process of migration towards an international destination, regardless of their legal status. We strongly encourage users to read more about our methodology before using this data online or in our data collection guidelines.</p>

```python
# Rename Columns Name for Easy Understandig
xdf = df.rename(columns = {'Main ID' : 'mainID', 'Incident ID' : 'incidentID', 'Region' : 'region',
                           'Incident Date' : 'incidentDate', 'Year' : 'year', 'Reported Month' : 'rptMonth', 
                           'Number Dead' : 'numDead', 'Minimum Estimated Number of Missing' : 'numMissing',
                           'Total Number of Dead and Missing' : 'totDeadMissing', 'Number of Survivors' : 'survivors', 
                           'Number of Females' : 'female', 'Number of Males' : 'male', 'Number of Children' : 'children',
                           'Cause of Death' : 'deathCause', 'Location of death' : 'deathLocation', 
                           'Information Source' : 'infoSource', 'Coordinates' : 'coordinates', 
                           'Migrantion route' : 'migrationRoute', 'URL' : 'url', 
                           'UNSD Geographical Grouping' : 'UNSDGeographGrouping', 'Source Quality' : 'sourceQuality'} )
xdf.columns
```

```python
#df.drop(labels=['Brand', 'Accel', 'TopSpeed'], axis=1, inplace=True)
xdf.drop(labels=['mainID', 'region', 'year', 'incidentDate','rptMonth', 'numDead', 'numMissing', 
                 'survivors', 'deathCause', 'deathLocation', 'infoSource', 'migrationRoute', 
                 'url', 'UNSDGeographGrouping', 'sourceQuality'],
                 axis=1, inplace=True)
xdf.columns
```

```python
# Replace the special character [,] from strong numbers.
# Fill Missing Values with Zero [0].
# Change Dataset from String to Integer.
# At Last Verify the Unique Values.
xdf['totDeadMissing'] #= xdf['totDeadMissing'].str
#xdf['totDeadMissing'].unique()
#xdf['totDeadMissing']
#replace(',', '', regex=True)
xdf['totDeadMissing'].fillna(0, inplace=True)
xdf['female'].fillna(0, inplace=True)
xdf['male'].fillna(0, inplace=True)
xdf['children'].fillna(0, inplace=True)
#xdf['totDeadMissing'] = xdf['totDeadMissing'].astype(int)
#xdf['totDeadMissing'].unique()
```

```python
totMigrants = [xdf['female'][i] + xdf['male'][i] + xdf['children'][i] for i in range(len(xdf))]
```

```python
xdf.insert(2, 'totalMigrants', totMigrants, True)
xdf.head()
```

```python
xdf.drop(labels=['female', 'male', 'children'], axis=1, inplace=True)
xdf.head()
```

```python
xCord = []
yCord = []
for cord in xdf['coordinates']:
    # delete comma
    cord = cord.replace(',','')
    # split string
    x, y = cord.split()
    # add x and y to lists
    xCord.append(float(x))
    yCord.append(float(y))
```

```python
xdf.insert(4, 'xCoord', xCord, True)
xdf.insert(5, 'yCoord', yCord, True)
xdf.head()
```

```python
xdf.drop(labels=['coordinates'], axis=1, inplace=True)
xdf.head()
```

```python
xdf.to_csv("missing_migrants_2022.csv", index=False)
```
